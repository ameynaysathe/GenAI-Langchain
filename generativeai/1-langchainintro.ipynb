### GROQ Model Integration

```python
from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

model = init_chat_model("groq:llama-3.1-8b-instant")
response = model.invoke("Give me a joke about machine learning.")
print(response.content)
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model="qwen/qwen3-32b")
response = model.invoke("Why do parrots talk?")
print(response.content)
```

### Streaming And Batch

```python
for chunk in model.stream("Write a easy on comeback"):
    print(chunk.text, end="", flush=True)
```

### Batch

```python
responses = model.batch(
    [
        "What is the capital of India?",
        "Who won ipl 2024?",
        "Give me a joke about machine learning.",
    ],
    config = {
        'max_concurrency': 5  # Limit the number of concurrent requests to 5    
    }
)
for response in responses:
    print(response)
```
